```python
# check how many have elements of orihinal title column
df.original_title.nunique()
```

```python
# analyze and drop Rows/Columns with Null values 
df = df.dropna()
```

```python
# display data frame   dimensions
print('Data columns\t\tDatatype')
print('â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“')
print(df.dtypes)
print('â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“')
print("Dimensions of the dataframe: ({}) rows and ({}) columns".format(df.shape[0],df.shape[1]))
```

```python
# pritn number of duplicates in TMDB movies datasets
print(df.duplicated().sum())
```

```python
# drop duplicates TMDB movies datasets
df.drop_duplicates(inplace=True)
```

```python
# drop columns unusing
df.drop(['imdb_id', 'keywords', 'overview', 'revenue_adj', 'budget_adj', 'homepage', 'vote_average', 'tagline'], axis=1, inplace=True)


print("After dropping columns: ({}) rows | ({}) columns".format(df.shape[0],df.shape[1]))
```

```python
# check if there any missing values
print('DataColumns\t  MissingValues')
print('â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“')
print(df.isna().sum())
print('â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“â€“')
print("({}) rows and ({}) columns".format(df.shape[0],df.shape[1]))
```

```python
# display how many element which value equal Zero in budget column 
df[(df['budget'] == 0)].shape[0]
```

```python
# display how many element which value equal Zero in revenue column 
df[(df['revenue'] == 0)].shape[0]
```

```python
# to calculate out the profits of each movie will subtract from budget
df['profit'] = df['revenue'] - df['budget']
df.tail()
```

```python
# spliting each string in the Series by sep and return a frame
# of dummy/indicator variables
# then merge them an take the mean as an index occurs
# in more than one
genres_df = df.genres.str.get_dummies(sep='|')
df = df.merge(genres_df, left_index=True, right_index=True)
```

```python
# Convert 'release_date' column to datetime
df['release_date'] = pd.to_datetime(df['release_date'])
```

```python
# detecting missing values
print(df.isna().sum())
```

### Research Question 1 (What are ten movies Vs highest profit for each one?)

```python
# ordering table by profit with only the specific columns
data = df.groupby(['profit', 'original_title'], as_index=False)['id', 'popularity', 'revenue', 'runtime', 'release_year'].sum()
data.tail(10)
```

```python
# arranging by highest value
counts = data.groupby(['profit', 'original_title'], as_index=False)['runtime'].sum().tail(10)
counts = counts.sort_values(by='profit', ascending = False)
counts
```

```python
# grouping a DataFrame by a column then each unique value
count = counts.groupby('profit')['original_title'].sum()
```

```python
# creating a bar chart for profit and orginal title
df = pd.DataFrame(count.index, count.values);
ax = df.plot.barh(rot=0)

ax.set_title('10 Most Profitable each of movie')
ax.set_xlabel('Profit')
ax.set_ylabel('Orginal Title');
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False);
ax.spines['left'].set_visible(False)
```

> We note `Avatar` movie is top one of profit where is more than billion and half.

### Research Question 2  (What is year has the muchest recording of movies?)

```python
# make group for each year and count the number of
# movies in each year
count_movie = df.groupby('release_year')['id'].count()
print(count_movie.tail(10))
```

```python
# creating a graph chart for number of movies vs each year
fig,ax= plt.subplots(figsize=(5, 5));
ax.plot(count_movie.index, count_movie.values); # data.index : x , data.values : y

ax.set_title('Year Vs Number Of Movies')
ax.set_xlabel('Relese Year')
ax.set_ylabel('Number Of Movies');
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```
